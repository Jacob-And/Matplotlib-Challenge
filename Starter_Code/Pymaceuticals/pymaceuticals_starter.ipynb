```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(study_results, mouse_metadata, on= "Mouse ID", how= "left")
# Display the data table for preview
merged_data.head()
```

```python
# Checking the number of mice.
merged_data['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
repeat_mouse = merged_data.loc[merged_data.duplicated(subset = 
    ["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
print(repeat_mouse)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
repeated_info_df = merged_data[merged_data["Mouse ID"] == "g989"]
repeated_info_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merged_data[merged_data["Mouse ID"].isin(repeat_mouse) == False]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
# Calculate summary statistics by treatment and display results
means = clean_df.groupby("Drug Regimen").mean(numeric_only=True)["Tumor Volume (mm3)"]

medians = clean_df.groupby("Drug Regimen").median(numeric_only=True)["Tumor Volume (mm3)"]

variances = clean_df.groupby("Drug Regimen").var(numeric_only=True)["Tumor Volume (mm3)"]

sds = clean_df.groupby("Drug Regimen").std(numeric_only=True)["Tumor Volume (mm3)"]

sems = clean_df.groupby("Drug Regimen").sem(numeric_only=True)["Tumor Volume (mm3)"]

summary_df = pd.DataFrame({"Mean Tumor Volume": means, 
                              "Median Tumor Volume": medians, 
                              "Tumor Volume Variance": variances, 
                              "Tumor Volume Std. Dev.": sds, 
                              "Tumor Volume Std. Err.": sems})
summary_df
```

## Bar and Pie Charts

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
test_mouse = clean_df.loc[clean_df["Mouse ID"] == "r944"]

timepoint = test_mouse['Timepoint']
volume = test_mouse['Tumor Volume (mm3)']
plt.plot(timepoint, volume)
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Amount of Capomulin Treatment for Mouse r944")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]

capomulin_mean = capomulin_data.groupby("Mouse ID").mean(numeric_only=True).reset_index()

mouse_weight = capomulin_mean["Weight (g)"]
mean_volume = capomulin_mean["Tumor Volume (mm3)"]

plt.scatter(mouse_weight, mean_volume)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Observed Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(mouse_weight, mean_volume)
print(correlation)

x_values = mouse_weight
y_values = mean_volume
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values*slope + intercept
plt.scatter(x_values, y_values)
plt.plot(x_values, regress_values, "r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
